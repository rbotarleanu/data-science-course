```python
# The prices might differ; update accordingly
corrected_prices = 0
for i, apt in enumerate(data):
    if apt['price'] not in apt['title']:
        splitted_title = apt['title'].split('-')
        print(splitted_title)
        for elem in splitted_title:
            if 'EUR' in elem:
                elem = elem.strip().split(' ')
                new_price = [e for e in elem if e != 'EUR'][0]
                data[i]['price'] = new_price
                corrected_prices += 1
```

```python
def floor_str_to_floor_int(floor):
    if floor != floor:
        return 0, 0
    all_floors = re.findall('(\d+)', floor)
    actual_floor = 0
    total_floors = 0
    if len(all_floors) == 1:
        try:
            total_floors = all_floors[0]
            if 'mansarda' in floor.lower():
                actual_floor = int(total_floors) + 1
        except:
            pass
    elif len(all_floors) == 2:
        actual_floor = int(all_floors[0])
        total_floors = int(all_floors[1])
    return actual_floor, total_floors
```

```python
af = []
tf = []
for floor in list(df.floor):
    actual_floor, total_floors = floor_str_to_floor_int(floor)
    af.append(actual_floor)
    tf.append(total_floors)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error

from sklearn.linear_model import LinearRegression
```

```python
X = np.array(df[regression_vars])
y = np.array(df['price'])
```

```python
X.shape, y.shape
```

```python
models = []
```

```python
kfold = KFold(n_splits=5)
for split, (train_indices, test_indices) in enumerate(kfold.split(range(X.shape[0]))):
    train_X, train_y = X[train_indices], y[train_indices]
    test_X, test_y = X[test_indices], y[test_indices]
    
    model = LinearRegression(normalize=True)
    model.fit(train_X, train_y)
    y_pred = model.predict(test_X)
    score = np.sqrt(mean_squared_error(test_y, y_pred))
    models.append(model)
    print('%d - score: %0.2f' % (split, score))
```
