# Conversion functions for various columns

```python
def parse_commision(commision):
    if commision.strip() == '':
        return float('nan')
    try:
        return re.match(r'\D*(\d+)\D*', commision).group(1)
    except:
        return float('nan')
```

```python
def parse_number(value):
    try:
        return float(value)
    except:
        return float('NaN')
```

```python
def parse_surface(char):
    char = char.replace('.', '')
    char = char.replace(',', '.')
    if 'mp' in char:
        try:
            return float(re.match(r'\D*(\d+\.?\d+)\s*mp.*', char).group(1))
        except:
            pass
    return float('NaN')
```

```python
def parse_year(char):
    try:
        return int(re.match(r'\D*(\d+)\D*', char).group(1))
    except:
        return float('nan')
```

```python
def parse_floor(char):
    if 'parter' in char.lower():
        floor = 0
    elif 'etaj' in char.lower():
        try:
            return int(re.match(r'.*etaj\s*(\d+)', char.lower()).group(1))
        except:
            return float('nan')
```

```python
def map_compartimentalization(comp):
    try:
        return {
        'decomandat': 0,
        'semidecomandat': 1,
        'nedecomandat': 2,
        'circular': 3,
        'vagon': 4
        }[comp]
    except:
        return 5
```

```python
def map_comfort(comf):
    if comf == 'lux':
        return 4
    return float(comf)
```

```python
def map_balconies(balc):
    if type(balc) is float:
        return balc
    if 'închise' not in balc:
        return float(balc)
    else:
        return -int(re.match(r'.*(\d+).*închise', balc).group(1))
```

```python
def map_structure(structure):
    try:
        return {
            'caramida': 1,
            'altele': 0,
            'metal': 2,
            'bca': 3,
            'beton': 4,
            'lemn': 5
        }[structure]
    except Exception as e:
        return float('nan')
```

```python
def map_type(t):
    if t == 'bloc de apartamente':
        return 1
    elif t == 'casa/vila':
        return 2
    else:
        return 0
```

```python
def parse_price(price):
    price = price.replace('.', '')
    price = price.replace(',', '')
    try:
        return float(price)
    except:
        return float('nan')
```

```python
def parse_dist_underground(dist_underground):
    try:
        return int(re.match(r'.*\((\d+).*minut', dist_underground).group(1))
    except:
        return float('nan')
```

# Constructing the DataFrame

```python
data = []
with open('dataset.txt', 'rt') as fin:
    for line in fin:
        d = json.loads(line)
        if 'An construcţie:' in d:
            d['An construcţie:'] = parse_year(d['An construcţie:'])
        if 'Etaj:' in d:
            d['Etaj:'] = parse_floor(d['Etaj:'])
        if 'Suprafaţă construită:' in d:
            d['Suprafaţă construită:'] = parse_surface(d['Suprafaţă construită:'])
        if 'Suprafaţă utilă totală:' in d:
            d['Suprafaţă utilă totală:'] = parse_surface(d['Suprafaţă utilă totală:'])
        if 'Suprafaţă utilă:' in d:
            d['Suprafaţă utilă:'] = parse_surface(d['Suprafaţă utilă:'])
        del d['currency']
        data.append(d)
```

# Model creation (no location data, only text mined data)

```python
# We select only variables linearly correla|ted with the price|
independent_variables = ['comfort', 'bathrooms', 'rooms', 'useful surface']
independent_variables = ['year', 'compartimentalization', 'comfort', 'floor', 'balconies',
                        'kitchens', 'bathrooms', 'rooms', 'garages', 'parking spots', 'structure',
                        'built surface', 'total useful surface', 'useful surface', 'type',
                        'dist_underground']
```

```python
# Build and evaluate a model using 5-fold CV
X = np.array(df[independent_variables])
y = np.array(df['price'])

scores = []

kfold = KFold(n_splits=5, random_state=42)

for split, (train_indices, test_indices) in enumerate(kfold.split(range(X.shape[0]))):
    train_X, train_y = X[train_indices], y[train_indices]
    test_X, test_y = X[test_indices], y[test_indices]

    
    scaler = StandardScaler()

    # Add text mined data
    vectorizer = TfidfVectorizer(min_df=.2, max_df=0.8)
    train_text_data = vectorizer.fit_transform(df['description'].reindex(train_indices))
    test_text_data = vectorizer.transform(df['description'].reindex(test_indices))
    
    train_X = np.hstack([train_X, train_text_data.todense()])
    test_X = np.hstack([test_X, test_text_data.todense()])

    # Scale the features using z-score scaling (subtract the mean, divide by std)
    train_X = scaler.fit_transform(train_X)
    test_X = scaler.transform(test_X)

    model = LinearRegressionTF(reg=.05)
    model.fit(train_X, train_y)
    y_pred = model.predict(test_X)
    score = np.sqrt(mean_squared_error(test_y, y_pred))  # RMSE
    print('[Split %d] Score: %f' % (split, score))
    scores.append(score)
```

```python
pd.Series(scores).describe()
```

# Model creation (location one-hot encondigs)

```python
# We select only variables linearly correlated with the price|
independent_variables = ['comfort', 'bathrooms', 'rooms', 'useful surface']
independent_variables = ['year', 'compartimentalization', 'comfort', 'floor', 'balconies',
                        'kitchens', 'bathrooms', 'rooms', 'garages', 'parking spots', 'structure',
                        'built surface', 'total useful surface', 'useful surface', 'type',
                        'dist_underground']
```

```python
# Build and evaluate a model using 5-fold CV
X = np.array(df[independent_variables])
y = np.array(df['price'])

scores = []

# This induces look-ahead bias but the ratio of samples to unique locations is too high
location_vectorizer = OneHotEncoder()
location_vectorizer.fit(df['location'].ravel().reshape(-1, 1))

kfold = KFold(n_splits=5, random_state=42)

for split, (train_indices, test_indices) in enumerate(kfold.split(range(X.shape[0]))):
    train_X, train_y = X[train_indices], y[train_indices]
    test_X, test_y = X[test_indices], y[test_indices]

    scaler = StandardScaler()

    # Add text mined data
    vectorizer = TfidfVectorizer(min_df=.2, max_df=0.8)
    train_text_data = vectorizer.fit_transform(df['description'].reindex(train_indices))
    test_text_data = vectorizer.transform(df['description'].reindex(test_indices))
    
    train_X = np.hstack([train_X, train_text_data.todense()])
    test_X = np.hstack([test_X, test_text_data.todense()])

    # Scale the features using z-score scaling (subtract the mean, divide by std)
    train_X = scaler.fit_transform(train_X)
    test_X = scaler.transform(test_X)

    # Add the location one-hot encodings
    train_location_data = location_vectorizer.transform(
        df['location'].reindex(train_indices).ravel().reshape(-1, 1))
    test_location_data = location_vectorizer.transform(
        df['location'].reindex(test_indices).ravel().reshape(-1, 1))
    
    train_X = np.hstack([train_X, train_location_data.todense()])
    test_X = np.hstack([test_X, test_location_data.todense()])

    # Train the model
    
    model = LinearRegressionTF(reg=.05) # needed since the one-hot encodings make the X.t%*%X matrix degenerate
    model.fit(train_X, train_y)
    y_pred = model.predict(test_X)
    score = np.sqrt(mean_squared_error(test_y, y_pred))  # RMSE
    print('[Split %d] Score: %f' % (split, score))
    scores.append(score)
```

```python
pd.Series(scores).describe()
```

```python
# Add text mined data and the location embeddings
vectorizer = TfidfVectorizer(min_df=.2, max_df=0.8)
text_data = vectorizer.fit_transform(df['description'])
X = np.hstack([X, text_data.todense()])
scaler = StandardScaler()
X = scaler.fit_transform(X)
location_data = location_vectorizer.transform(df['location'].ravel().reshape(-1, 1)) 
X = np.hstack([X, location_data.todense()])
```

```python
lr = LinearRegressionTF(reg=.05)
lr.fit(X, y)
```

```python
np.sqrt(mean_squared_error(lr.predict(X), y))
```
